```python
import pandas as pd
import numpy as np
import xgboost as xgb
import pickle

# ---------------- Load data ----------------
df = pd.read_csv("train_prices_decisions_2025.csv")

# Features: covariates + price
X = df[['Covariate1','Covariate2','Covariate3','price_item']]
y = df['item_bought'].astype(int)

# ---------------- Train XGBoost on entire dataset ----------------
model = xgb.XGBClassifier(
    max_depth=5,
    n_estimators=1000,
    use_label_encoder=False,
    eval_metric='logloss',
    n_jobs=-1,
    random_state=42
)
model.fit(X, y)

# ---------------- Sample 2000 customer covariates ----------------
num_samples = 4000
sample_covariates = df[['Covariate1','Covariate2','Covariate3']].sample(num_samples, random_state=42).values

# ---------------- Save model + sample covariates as pickle ----------------
save_dict = {
    'model': model,
    'sample_covariates': sample_covariates
}

with open("xgb_model", "wb") as f:
    pickle.dump(save_dict, f)

print("Saved XGBoost model and sampled covariates to xgb_model_and_samples.pkl")
```
